> ### What is Exploratory Data Analysis?
> **Exploratory data analysis (EDA)** is a technique used by data scientists to inspect, characterize and briefly summarize the contents of a dataset. EDA is often the first step when encountering a new or unfamiliar dataset. EDA helps the data scientist become acquainted with a dataset and test some basic assumptions about the data. By the end of the EDA process, some initial insights can be drawn from the dataset and a framework for further analysis or modeling is established.

## 3. Summarize the Dataset

```python
# sumarize the dataset by calculating some basic summary statisitcs

df.describe()
```

```python
# group the dataset by average student-teacher ratio by school district

ratio = df.groupby(['Community School District']).mean()
```

```python
# list the average student-teacher ratio in each school district

ratio
```

```python
# group by the highest student teacher ratio per district

max = df.groupby(['Community School District']).max()
```

```python
# list the max student-teacher ratio in each school district

max
```

```python
sort1 = df.sort_values('School Pupil-Teacher Ratio', ascending = False)
```

```python
sort1.tail(5)
```

```python
sort1.head(5)
```

```python
# group by the lowest student teacher ratio per district

min = df.groupby(['Community School District']).min()
```

```python
# list the min student-teacher ratio in each school district

min
```

## 4. Visualize the Dataset

```python
plt.figure(figsize=(15, 20))
sns.barplot(x='School Pupil-Teacher Ratio' ,y='Community School District',color='#4f81bd', saturation=.85, ci=None,data=df)
```

```python
plt.figure(figsize=(15, 20))
ax = sns.boxplot(y=df['Community School District'],x=df['School Pupil-Teacher Ratio'], color='#4f81bd', saturation=.85)
```
